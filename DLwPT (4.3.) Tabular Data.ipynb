#Experiment with Wine Data
The data contains chemical characterizations of samples of *vihno verde* wine from Portugal, that also includes sensory quality score of those wines.

**DOWNLOAD SOURCE**:
https://raw.githubusercontent.com/deep-learning-with-pytorch/dlwpt-code/master/data/p1ch4/tabular-wine/winequality-white.csv

```
import csv
wine_path="/content/raw.githubusercontent.com/deep-learning-with-pytorch/dlwpt-code/master/data/p1ch4/tabular-wine/winequality-white.csv"
wineq_numpy=np.loadtxt(wine_path,dtype=np.float32,delimiter=";",skiprows=1)
wineq_numpy
```

The data consists of 12 columns where the last column 'quality' contains the sensory quality score from 0 (very bad) to 10 (exellent) and the other 11 columns contain values of chemical variables.

```
col_list=next(csv.reader(open(wine_path),delimiter=';'))
wineq_numpy.shape,col_list
```

Using machine learning algorithm, we are going to predict the quality of wine based on its chemical characteristics.

```
#Converting the data which came in the form of NumPy array to PyTorch tensor
wineq=torch.from_numpy(wineq_numpy)
wineq.shape,wineq.dtype
```

```
data=wineq[:,:-1] #Selecting all rows, and all columns except the last one
data,data.shape
```

```
target=wineq[:,-1] #Selecting all rows, and the last column only, as the target
target,target.shape
```

```
target=wineq[:,-1].long() #Long is a data type similat to Integer but for higher value
target
```

```
target_oneshot=torch.zeros(target.shape[0],10)
target_oneshot.scatter_(1,target.unsqueeze(1),1.0)
```

```
#'unsqueeze' returns a new tensor with a dimension of size one inserted at the specified position.
#The returned tensor shares the same underlying data with this tensor.
target_unsqueezed=target.unsqueeze(1)
target_unsqueezed
```

```
data_mean=torch.mean(data,dim=0)
data_mean
```

```
data_var=torch.var(data,dim=0)
data_var
```

```
data_normalized=(data-data_mean)/torch.sqrt(data_var)
data_normalized
```

```
bad_indexes=target<=3 
#PyTorch also provides comparisons functions
#Here torch.le(target,3), but using operators seems to be a good standard
bad_indexes.shape,bad_indexes.dtype,bad_indexes.sum()
```

```
bad_data=data[bad_indexes]
bad_data.shape
```

```
bad_data=data[target<=3]
mid_data=data[(target>3)&(target<7)]
#For boolean numpy array and PyTorch tensors, the & operator does a logical "and" operation
good_data=data[target>=7]

bad_mean=torch.mean(bad_data,dim=0)
mid_mean=torch.mean(mid_data,dim=0)
good_mean=torch.mean(good_data,dim=0)

for i, args in enumerate(zip(col_list,bad_mean,mid_mean,good_mean)):
  print('{:3} {:22} {:8.2f} {:8.2f} {:8.2f}'.format(i,*args))
```

```
total_sulfur_threshold=141.83
total_sulfur_data=data[:,6]
predicted_indexes=torch.lt(total_sulfur_data,total_sulfur_threshold)

predicted_indexes.shape,predicted_indexes.dtype,predicted_indexes.sum()
```

```
actual_indexes=target>5
actual_indexes.shape,actual_indexes.dtype,actual_indexes.sum()
```

```
n_matches=torch.sum(actual_indexes & predicted_indexes).item()
n_predicted=torch.sum(predicted_indexes).item()
n_actual=torch.sum(actual_indexes).item()

n_matches,n_matches/n_predicted,n_matches/n_actual
```
